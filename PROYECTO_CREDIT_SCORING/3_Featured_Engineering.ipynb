# SALDOS ACTIVOS

```python
Ind_TDC = Base_Independientes_A[Base_Independientes_A['SECTOR1'] == 'TDC']
Ind_HIPO = Base_Independientes_A[Base_Independientes_A['SECTOR1'] == 'HIPOTECA RESIDENCIAL']
Ind_CONS = Base_Independientes_A[Base_Independientes_A['SECTOR1'] == 'CONSUMO PERSONAL']
Ind_AUTO = Base_Independientes_A[Base_Independientes_A['SECTOR1'] == 'AUTO']
```

```python
Saldo_TDC = Ind_TDC.groupby(['mes','CLIENTE'])['SALDO'].agg({'SALDO_TDC_UM':"sum"}).reset_index()
Saldo_HIPO = Ind_HIPO.groupby(['mes','CLIENTE'])['SALDO'].agg({'SALDO_HIPO_UM':"sum"}).reset_index()
Saldo_CONS = Ind_CONS.groupby(['mes','CLIENTE'])['SALDO'].agg({'SALDO_CONS_UM':"sum"}).reset_index()
Saldo_AUTO = Ind_AUTO.groupby(['mes','CLIENTE'])['SALDO'].agg({'SALDO_AUTO_UM':"sum"}).reset_index()
```

```python
Saldo_TDC.head(10)
```

```python
SALDOS_ACT = pd.merge(Saldo_TDC, Saldo_CONS, on=['mes','CLIENTE'], how='outer')
SALDOS_ACT = pd.merge(SALDOS_ACT, Saldo_HIPO, on=['mes','CLIENTE'], how='outer')
SALDOS_ACT = pd.merge(SALDOS_ACT, Saldo_AUTO, on=['mes','CLIENTE'], how='outer')
 
SALDOS_ACT
```

```python
SALDOS_ACT.describe()
```

```python
date_format = '%Y-%m'
SALDOS_ACT['mes'] = SALDOS_ACT['mes'].apply(lambda x: datetime.strptime(str(x),date_format).date())
SALDOS_ACT['mes'] = SALDOS_ACT['mes'] + relativedelta(months=1)
SALDOS_ACT['mes'] = SALDOS_ACT['mes'].astype(str)
SALDOS_ACT['mes_inicio'] = SALDOS_ACT.mes

del SALDOS_ACT['mes']
```

```python
SALDOS_ACT
```

#### CALCULAR ANTIGUEDAD DE SU ULTIMO EMPLEO

```python
base_modelo['antiguedad_empleo'] = -(base_modelo['FechaEmpleoActual'] - base_modelo['mes_inicio'])
base_modelo['antiguedad_empleo'] = base_modelo['antiguedad_empleo'].apply(lambda x: x.days/365)
base_modelo['antiguedad_empleo'] = base_modelo['antiguedad_empleo'].apply(lambda x: x if x<100 else np.nan)
base_modelo['antiguedad_empleo'] = base_modelo['antiguedad_empleo'].apply(lambda x: x if x>=0 else np.nan)
```
