```python
import pandas as pd
from datetime import datetime
pd.set_option('display.max_columns', 500)
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
Base_Independientes = pd.read_excel('../data/raw/Base_Independientes.xlsx')
```

```python
Base_Independientes.SECTOR1 = Base_Independientes.SECTOR1.str.upper()
Base_Independientes[Base_Independientes['cartera'] == 'activo']['SECTOR1'].value_counts()
```

```python
date_format = '%Y%m%d'
Base_Independientes['FECHA'] = Base_Independientes.FECHA.apply(lambda x: datetime.strptime(str(x),date_format))
Base_Independientes['FECHA_INICIO'] = Base_Independientes.FECHA_INICIO.apply(lambda x: datetime.strptime(str(x),date_format))
```

```python
filtro = (Base_Independientes['SECTOR1'] == 'TDC') |  (Base_Independientes['SECTOR1'] == 'HIPOTECA RESIDENCIAL') | (Base_Independientes['SECTOR1'] == 'CONSUMO PERSONAL')| (Base_Independientes['SECTOR1'] == 'AUTO') 
```

```python
Base_Independientes_f = Base_Independientes[filtro]
Base_Independientes_f = Base_Independientes_f.reset_index(drop=True)
```

```python
Base_Independientes_f['MADUREZ'] = Base_Independientes_f['FECHA'] - Base_Independientes_f['FECHA_INICIO']
Base_Independientes_f['MADUREZ'] = Base_Independientes_f['MADUREZ'].apply(lambda x: math.ceil(x.days/30))
```

```python
base_objetivo['Fecha_Mora30'] = base_objetivo['max'].apply(lambda x: 1 if x>=30 else None)
base_objetivo['Fecha_Mora60'] = base_objetivo['max'].apply(lambda x: 1 if x>=60 else None)
base_objetivo['Fecha_Mora90'] = base_objetivo['max'].apply(lambda x: 1 if x>=90 else None)
```

```python
for i in base_objetivo.index:
    
    if base_objetivo.loc[i,'Fecha_Mora30'] == 1:
        base_objetivo.loc[i,'Fecha_Mora30'] = str(base_objetivo.loc[i,'mes'])[:4] + str(base_objetivo.loc[i,'mes'])[5:]
        base_objetivo.loc[i,'Fecha_Mora30'] = int ( base_objetivo.loc[i,'Fecha_Mora30'])
```

```python
for i in base_objetivo.index:
    
    if base_objetivo.loc[i,'Fecha_Mora90'] == 1:
        base_objetivo.loc[i,'Fecha_Mora90'] = str(base_objetivo.loc[i,'mes'])[:4] + str(base_objetivo.loc[i,'mes'])[5:]
        base_objetivo.loc[i,'Fecha_Mora90'] = int ( base_objetivo.loc[i,'Fecha_Mora90'])
```
